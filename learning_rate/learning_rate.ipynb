```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

def f(x):
    return x**2

def df(x):
    return 2*x

def gradient_descent(x_start, learning_rate, epochs):
    x = x_start
    history = [x]

    for i in range(epochs):
        x = x - learning_rate * df(x)
        history.append(x)
    return history

x_start = 2.5
epochs = 30

learning_rates = [0.01, 0.1, 0.9]

sns.set_theme(style="ticks", font_scale=1.7)
x = np.linspace(-3, 3, 400)
y = f(x)

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for ax, lr in zip(axs, learning_rates):
    history = gradient_descent(x_start, lr, epochs)
    ax.plot(x, y, 'r')
    ax.plot(history, [f(x) for x in history], 'o-', label=f"ε={lr}")
    
    ax.plot([], [], ' ', label=f'Steps={epochs}')
    
    ax.legend()
    ax.set_xlim([-2, 2])
    ax.set_ylim([-0.5, 3])  # Adjust the y-axis limits here
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.grid()

# Display the figure
plt.tight_layout()
plt.savefig("learning_rate.pdf", format="pdf", dpi=300, bbox_inches='tight')
```
