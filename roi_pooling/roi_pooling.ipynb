```python
import seaborn as sns
from random import randint
import matplotlib.pyplot as plt
```

```python
def randbetween(x, y):
    return randint(x*100, y*100)/100
```

```python
data = []
for i in range(6):
    data.append([])
    for j in range(6):
        data[i].append(randbetween(0, 1))
```

```python
sns.set_theme(font_scale=1.3, font="")
sns.heatmap(data, annot=True, yticklabels=False, xticklabels=False, cbar=False, square=True, cmap="viridis")
plt.tight_layout()
plt.savefig("heatmap.svg", format="svg", dpi=300, bbox_inches='tight')
```

```python
data = [[0.91, 0.73], [0.97, 0.51]]
sns.set_theme(font_scale=3.3, font="")
sns.heatmap(data, annot=True, yticklabels=False, xticklabels=False, cbar=False, square=True, cmap="viridis", vmin=0, vmax=1)
plt.tight_layout()
plt.savefig("heatmap_detail.svg", format="svg", dpi=300, bbox_inches='tight')
```
