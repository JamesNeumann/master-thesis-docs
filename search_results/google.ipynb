```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
sb.set_theme(style="ticks")

# sb.set()
```

```python
df = pd.read_csv("./google.csv")
df
```

```python
fig = plt.figure(figsize=(15, 6))
# sb.set(font_scale=2, style="ticks")
sb.set_theme("poster")
sb.set_style("ticks")
sb.lineplot(data=df, x='Woche', y='TensorFlow', label="TensorFlow")
sb.lineplot(data=df, x='Woche', y='PyTorch', label="PyTorch")
sb.lineplot(data=df, x='Woche', y='Keras', label="Keras")
sb.lineplot(data=df, x='Woche', y='Caffe', label="Caffe")
sb.lineplot(data=df, x='Woche', y='MXNet', label="MXNet")
# sb.lineplot(data=df)
# fig.legend()  # Add legend labels for each lineplot
# Set the y-axis tick label font size
# plt.setp(gfg.get_legend().get_title(), fontsize='16')
# plt.title('Google Search trends')
# plt.xlabel('Date')
plt.ylabel('Relative search popularity in %')
plt.xticks(rotation=0, ha='center')  # Rotate x-axis labels and align them to the right
plt.gca().xaxis.set_major_locator(ticker.AutoLocator())  # Specify the desired number of x-axis labels (e.g., 6)
plt.tight_layout()
plt.xlabel("")
plt.savefig("google.pdf", format="pdf", dpi=300, bbox_inches='tight')
```
